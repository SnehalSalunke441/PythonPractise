```python
import pandas as pd
iris = pd.read_csv('iris.csv', names=['sepal_length','sepal_width' ,'petal_length', 'petal_width', 'class'])
iris.head()
```

```python
# Matplotlib scatterplot

import matplotlib.pyplot as plt

fig, ax=plt.subplots()
ax.scatter(iris['sepal_length'], iris['sepal_width'])

ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
# Matplotlib Scatterplot coloured by class

colors={'Iris-setosa' :'r', 'Iris-versicolor': 'g', 'Iris-virginica': 'b'}

fig, ax=plt.subplots()

for i in range(len(iris['sepal_length'])):
    ax.scatter(iris['sepal_length'][i], iris['sepal_width'][i], color = colors[iris['class'][i]])
    
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')  
```

```python
# Line Plot

import numpy as np

df = pd.DataFrame({'x': range(1,11), 'y':np.random.randn(10)})
plt.plot('x', 'y', data=df, color='blue')
plt.show()
```

```python
# Lineplot 2

columns = iris.columns.drop(['class'])
x_data =range(0, iris.shape[0])
fig, ax =plt.subplots()

for column in columns:
    ax.plot(x_data, iris[column], label ='column')
    
ax.set_title('Iris Dataset')
ax.legend()
```

```python
# Histogram

fig, ax =plt.subplots()

ax.hist(iris['sepal_width'])

ax.set_title('Sepal Width')
ax.set_xlabel('Width')
ax.set_ylabel('Frequency')
```

```python
# Bar chart

height =[3, 12, 5, 18, 45]
bars=('A', 'B', 'C', 'D', 'E')

y_pos =np.arange(len(bars))

plt.bar(y_pos, height)
plt.xticks(y_pos, bars)
plt.show()
```

```python
# Bar Chart

fig, ax =plt.subplots()

data=iris['class'].value_counts()

names= data.index
frequency =data.values

ax.bar(names, frequency)
ax.set_title('Iris Data')
ax.set_xlabel('Names')
ax.set_ylabel('Frequency')
```

```python
# ScatterPlot
iris.plot.scatter(x='sepal_length', y='sepal_width', title='Iris Dataset')
```

```python
# Line Chart
iris.drop(['class'], axis=1).plot.line(title='Iris Dataset')
```
